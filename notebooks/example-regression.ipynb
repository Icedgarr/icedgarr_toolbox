```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
data_dict = datasets.load_boston()
data = pd.DataFrame(data_dict['data'], columns=data_dict['feature_names'])
data['target'] = data_dict['target']
```

```python
train, test = train_test_split(data)
```

```python
feature_columns = data_dict['feature_names']
```

```python
train
```

### Exploration

```python

from icedgarr_toolbox.exploration import plot_corr_matrix, compare_train_test_distributions
from icedgarr_toolbox.exploration import plot_features_histograms, train_test_classification_score
```

```python
plot_corr_matrix(train)
```

```python
plot_features_histograms(train_data=train, test_data=test, features=feature_columns, bins=50)
```

```python
# High p-value -> the distributions are probably similar. Low p-value -> the distributions are probably different
compare_train_test_distributions(train, test, feature_columns)
```

```python
# Close to 0.5 or less means the two sets are not distinguishable.
predictions = train_test_classification_score(train, test, feature_columns)
```

### Feature engineering

```python

from icedgarr_toolbox.feature_engineering import compute_percentiles_list_columns, create_statistics
```

```python
# Compute the percentile of each value per column
features = compute_percentiles_list_columns(train, feature_columns)

#Compute percentiles for both train and test
train['train'] = 1
test['train'] = 0
train_and_test = pd.concat([train, test])
features = compute_percentiles_list_columns(train_and_test, feature_columns)
features_train = features.loc[train_and_test['train'] == 1]
features_test = features.loc[train_and_test['train'] == 0]
```

```python
features_train = features_train.join(create_statistics(features_train, columns=features_train.columns, 
                                                       prefix='percentile_features'))
```

### Modeling

```python
from sklearn.model_selection import KFold

from icedgarr_toolbox.metrics.regression.ermse_metric import ermse_metric
from icedgarr_toolbox.preprocessing import create_pca_embedding
from icedgarr_toolbox.modeling import MeanPredictor
from icedgarr_toolbox.modeling import CrossValidator
```

```python
pca = create_pca_embedding(train, 5)
```

```python
model = MeanPredictor()
model.fit(train, train['target'])
predictions = model.predict(test)
ermse_metric(test['target'], predictions)
```

```python
splitter = KFold(5)
cv = CrossValidator(splitter, ermse_metric)
results = cv.compute_scores(model, train, feature_columns=feature_columns, target_column='target')
```
